Step 1: Load Dataset
-

```python
import pandas as pd 

df = pd.read_csv(
    "/content/hiring_dataset.csv"
)

df.head()
```

Step 2: Select features
-

```python
features = [
    "Skills",
    "Experience (Years)",
    "Education",
    "Certification",
    "Salary Expectation ($)",
    "Projects Count"
]
```

```python
target = "Recruiter Decision"
```

Step 3: Encode Target
-

```python
from sklearn.preprocessing import LabelEncoder 

target_encoder = LabelEncoder()

df[target] = target_encoder.fit_transform(
    df[target]
)
```

Step 4: Encode Education
-

```python
edu_encoder = LabelEncoder()

df["Education"] = edu_encoder.fit_transform(
    df["Education"]
)
```

Step 5: Encode Certifications 
-

```python
cert_encoder = LabelEncoder()

df["Certifications"] = cert_encoder.fit_transform(
    df["Certifications"]
)
```

Step 6: TF-IDF for skills
-

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=100
)

skills_matrix = tfidf.fit_transform(
    df["Skills"]
)
```

```python
# convert to dataframe 

skills_df = pd.DataFrame(
    skills_matrix.toarray(),
    columns=tfidf.get
)
```
